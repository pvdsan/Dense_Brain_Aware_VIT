```python
import os
import shutil

# Define source and destination directories
source_dir = 'smri_working_memory_data'
destination_dir = 'all_precomputed_tensors'

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)

# Initialize a counter for sequential naming
counter = 0

# Iterate through each GPU directory in order
for gpu_id in range(4):  # Adjust the range if you have more or fewer GPUs
    gpu_dir = os.path.join(source_dir, f"gpu_{gpu_id}")
    print(gpu_dir)
    if os.path.exists(gpu_dir):
        # Get all subject directories in the current GPU directory
        subject_dirs = [d for d in os.listdir(gpu_dir) if os.path.isdir(os.path.join(gpu_dir, d))]
        
        for subject_dir in subject_dirs:
            src_path = os.path.join(gpu_dir, subject_dir)
            
            # Define the new subject directory name based on the counter
            new_subject_dir = f"subject_{counter}"
            dst_path = os.path.join(destination_dir, new_subject_dir)
            
            # Check for potential naming conflicts (shouldn't occur with sequential naming)
            if os.path.exists(dst_path):
                print(f"Destination path {dst_path} already exists. Skipping subject {subject_dir}.")
                counter += 1  # Increment counter to maintain sequence
                continue
            
            try:
                # Move and rename the subject directory
                shutil.move(src_path, dst_path)
                print(f"Moved {src_path} to {dst_path}")
                counter += 1  # Increment counter after successful move
            except Exception as e:
                print(f"Failed to move {src_path} to {dst_path}. Error: {e}")
                continue
    else:
        print(f"GPU directory {gpu_dir} does not exist. Skipping.")

# Summary
print("\nSummary:")
print(f"Total subjects moved and renamed: {counter}")
print(f"Destination directory: {destination_dir}")
```
